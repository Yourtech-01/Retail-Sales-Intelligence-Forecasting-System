```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/sample_data/clean_walmart_data.csv")
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.sort_values(['Store', 'Dept', 'Date'])
df.head()
```

```python
df['lag_1'] = df.groupby(['Store','Dept'])['Weekly_Sales'].shift(1)
df['lag_2'] = df.groupby(['Store','Dept'])['Weekly_Sales'].shift(2)
df['lag_4'] = df.groupby(['Store','Dept'])['Weekly_Sales'].shift(4)
```

```python
df['rolling_mean_4'] = (
    df.groupby(['Store','Dept'])['Weekly_Sales']
    .shift(1)
    .rolling(4)
    .mean()
)
```

```python
df['rolling_std_4'] = (
    df.groupby(['Store','Dept'])['Weekly_Sales']
    .shift(1)
    .rolling(4)
    .std()
)
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Week'] = df['Date'].dt.isocalendar().week.astype(int)
df['Quarter'] = df['Date'].dt.quarter
```

```python
df['IsHoliday'] = df['IsHoliday'].astype(int)
df['pre_holiday'] = (
    df.groupby(['Store','Dept'])['IsHoliday']
    .shift(-1)
    .fillna(0)
)
```

```python
markdown_cols = ['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5']

df['Total_Markdown'] = df[markdown_cols].sum(axis=1)
df['Avg_Markdown'] = df[markdown_cols].mean(axis=1)
```

```python
df = pd.get_dummies(df, columns=['Type'], drop_first=True)
df['Size_Log'] = np.log1p(df['Size'])
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.to_csv("/content/sample_data/feature_engineered_data.csv", index=False)
```
